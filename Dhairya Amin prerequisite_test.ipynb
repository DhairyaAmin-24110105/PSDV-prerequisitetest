Question 2:
Mutable vs Immutable types

Mutable data type can be changed within the code after its creation. For example: Lists,dictionaries and sets. We can change any element of these data types after their creation within the code. In lists, we can use list_name[index]=new value to change the specific element. In dictionaries, we can use dictionary_name[key]=new value to change the element.

Immutable data type cannot be changed or altered after its creation. It will reflect an error if we try to mainpulate the data within these data types. To make changes, we need to create another variable. For example: String and tuple are immutable data types. We cannot change the elements within these data types but we can create a new variable for making changes like s="hello" n=s[:3]+"t"+s[4:5](String slicing). The output is "helto" where "l" is replaced by "t" but it is only possible by creation of a new variable.

Question 3:
Fibonacci sequence

```python
import matplotlib.pyplot as plt
n=int(input())
l=[]
for i in range(n):
    if i==0:
        l.append(0)
    if i==1 or i==2:
        l.append(1)
    if i>2:
        l.append(l[i-1]+l[i-2])
print("First",n,"Fibonacci numbers:",l)
plt.plot(l)
plt.title("Fibonacci sequence")
plt.show()
```

Question 16:
Set Operations

```python
from matplotlib_venn import venn2
from matplotlib import pyplot as plt 
n=list(set(map(int,input("Set A:").split())))
k=list(set(map(int,input("Set B:").split())))
inter=[]
sym=[]
for i in n:
    if i in k:
        inter.append(i)
y=n+k
for i in inter:
    if i in y:
        y.remove(i)
for i in y:
    if i not in inter:
        sym.append(i)
s1=[]
s2=[]
for i in n:
    if i not in inter:
        s1.append(i)
for i in k:
    if i not in inter:
        s2.append(i)
print("Intersection:",set(inter))
print("Union:",set(y))
print("Symmetric difference:",set(sym))
diagram=venn2(subsets=(len(y)-len(k),len(y)-len(n),len(inter)),set_labels=("Set A","Set B"))
diagram.get_label_by_id("10").set_text(set(s1))
diagram.get_label_by_id("11").set_text(set(inter))
diagram.get_label_by_id("01").set_text(set(s2))
plt.show()
```

Question 21:
Basic Statistics

```python
import math
import matplotlib.pyplot as plt
n=list(map(int,input().split()))
n.sort()
avg=sum(n)/len(n)
print("Mean of numbers:",avg)
x=len(n)
def median(x):
    if x%2==0:
        u=(n[x//2]+n[(x//2)-1])/2
    else:
        u=n[(x-1)//2]
    return u
print("Median of numbers:",median(x))
t=list(set(n))
k=[]
for i in t:
    k.append(n.count(i))
if max(k)==1:
    print("No mode")
else:
    mode=t[k.index(max(k))]
    print("Mode of numbers is:",mode)
var=0
for i in n:
    var+=(i-avg)**2
var/=len(n)
sd=math.sqrt(var)
print("Standard deviation is:",sd)
plt.boxplot(n)
plt.title("Boxplot of Numbers in the list")
plt.show()
```

Question 22:
Normalization

```python
import matplotlib.pyplot as plt
n=list(map(int,input().split()))
def normalization(n):
    u=[]
    for i in n:
        l=(i-min(n))/(max(n)-min(n))
        l=round(l,2)
        u.append(l)
    return u
print("Normalized list:",normalization(n))
plt.plot(normalization(n))
plt.title("Normalization plot")
plt.show()
```
